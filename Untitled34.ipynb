```python
# Лабораторная работа №4 (Вариант 7)

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, callbacks

# Функция multi-hot кодирования
def multi_hot_encoding(reviews, num_words, remove_k_first=0):
    results = np.zeros((len(reviews), num_words))
    for i, review in enumerate(reviews):
        indices = [idx for idx in review if idx > remove_k_first]
        if indices:
            results[i, indices] = 1.0
    return results

# Этап 1. Бинарная классификация (IMDB)
print("ЭТАП 1. Бинарная классификация отзывов IMDB")

NUM_WORDS_IMDB = 15000
INDEX_FROM = 3

(train_data, train_labels), (test_data, test_labels) = keras.datasets.imdb.load_data(
    num_words=NUM_WORDS_IMDB, start_char=1, oov_char=2, index_from=INDEX_FROM)

train_features_imdb = multi_hot_encoding(train_data, NUM_WORDS_IMDB, remove_k_first=0)
test_features_imdb = multi_hot_encoding(test_data, NUM_WORDS_IMDB, remove_k_first=0)

# Модель по варианту 7: 3 слоя 16-64-64, активации tanh-relu-tanh
def create_imdb_v7():
    inputs = keras.Input(shape=(NUM_WORDS_IMDB,))
    x = layers.Dense(16, activation='tanh')(inputs)
    x = layers.Dense(64, activation='relu')(x)
    x = layers.Dense(64, activation='tanh')(x)
    outputs = layers.Dense(1, activation='sigmoid')(x)
    model = keras.Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='rmsprop',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

# Базовая модель: 2 слоя 16-16, активации relu-relu
def create_imdb_baseline():
    inputs = keras.Input(shape=(NUM_WORDS_IMDB,))
    x = layers.Dense(16, activation='relu')(inputs)
    x = layers.Dense(16, activation='relu')(x)
    outputs = layers.Dense(1, activation='sigmoid')(x)
    model = keras.Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='rmsprop',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

print("Обучение модели по варианту 7 ")
model_v7 = create_imdb_v7()
history_v7 = model_v7.fit(train_features_imdb, train_labels,
                          batch_size=512, epochs=20, verbose=2,
                          validation_split=0.2,
                          callbacks=[callbacks.EarlyStopping(patience=3, restore_best_weights=True)])

print(" Обучение базовой модели ")
baseline_imdb = create_imdb_baseline()
history_base_imdb = baseline_imdb.fit(train_features_imdb, train_labels,
                                      batch_size=512, epochs=20, verbose=2,
                                      validation_split=0.2,
                                      callbacks=[callbacks.EarlyStopping(patience=3, restore_best_weights=True)])

# Оценка на тесте
test_loss_v7, test_acc_v7 = model_v7.evaluate(test_features_imdb, test_labels)
test_loss_base, test_acc_base = baseline_imdb.evaluate(test_features_imdb, test_labels)
print(f"\nРезультаты на тестовой выборке:")
print(f"Вариант 7: accuracy = {test_acc_v7:.4f}")
print(f"Базовый классификатор: accuracy = {test_acc_base:.4f}")

# Графики
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history_v7.history['loss'], label='V7 train loss')
plt.plot(history_v7.history['val_loss'], label='V7 val loss')
plt.plot(history_base_imdb.history['loss'], '--', label='Base train loss')
plt.plot(history_base_imdb.history['val_loss'], '--', label='Base val loss')
plt.title('Функция потерь (IMDB)')
plt.legend()
plt.subplot(1,2,2)
plt.plot(history_v7.history['accuracy'], label='V7 train acc')
plt.plot(history_v7.history['val_accuracy'], label='V7 val acc')
plt.plot(history_base_imdb.history['accuracy'], '--', label='Base train acc')
plt.plot(history_base_imdb.history['val_accuracy'], '--', label='Base val acc')
plt.title('Точность (IMDB)')
plt.legend()
plt.tight_layout()
plt.show()

# Пример предсказания
sample_idx = 42
sample = test_features_imdb[sample_idx:sample_idx+1]
pred = model_v7.predict(sample)[0,0]
print(f"\nПример тестового отзыва {sample_idx}:")
print(f"Истинная метка: {test_labels[sample_idx]}")
print(f"Предсказание сети: {pred:.4f} ({'позитивный' if pred>0.5 else 'негативный'})")

# Этап 2. Многоклассовая классификация (Reuters)
print("ЭТАП 2. Многоклассовая классификация новостей Reuters")

NUM_WORDS_REUTERS = 10000
(train_data_r, train_labels_r), (test_data_r, test_labels_r) = keras.datasets.reuters.load_data(
    num_words=NUM_WORDS_REUTERS)

train_features_r = multi_hot_encoding(train_data_r, NUM_WORDS_REUTERS, remove_k_first=0)
test_features_r = multi_hot_encoding(test_data_r, NUM_WORDS_REUTERS, remove_k_first=0)

num_classes = int(np.max(train_labels_r)) + 1  # обязательно int
train_labels_cat = keras.utils.to_categorical(train_labels_r, num_classes)
test_labels_cat = keras.utils.to_categorical(test_labels_r, num_classes)

# Модель по варианту 7: 2 слоя 64-32, активации elu-relu
def create_reuters_v7():
    inputs = keras.Input(shape=(NUM_WORDS_REUTERS,))
    x = layers.Dense(64, activation='elu')(inputs)
    x = layers.Dense(32, activation='relu')(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    model = keras.Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='rmsprop',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model

# Базовая модель: 2 слоя 64-64, активации relu-relu
def create_reuters_baseline():
    inputs = keras.Input(shape=(NUM_WORDS_REUTERS,))
    x = layers.Dense(64, activation='relu')(inputs)
    x = layers.Dense(64, activation='relu')(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    model = keras.Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='rmsprop',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model

print("Обучение модели по варианту 7")
reuters_v7 = create_reuters_v7()
history_r7 = reuters_v7.fit(train_features_r, train_labels_cat,
                            batch_size=512, epochs=30, verbose=2,
                            validation_split=0.2,
                            callbacks=[callbacks.EarlyStopping(patience=4, restore_best_weights=True)])

print(" Обучение базовой модели")
reuters_base = create_reuters_baseline()
history_rbase = reuters_base.fit(train_features_r, train_labels_cat,
                                 batch_size=512, epochs=30, verbose=2,
                                 validation_split=0.2,
                                 callbacks=[callbacks.EarlyStopping(patience=4, restore_best_weights=True)])

# Оценка на тесте
test_loss_r7, test_acc_r7 = reuters_v7.evaluate(test_features_r, test_labels_cat)
test_loss_rbase, test_acc_rbase = reuters_base.evaluate(test_features_r, test_labels_cat)
print(f"\nРезультаты на тестовой выборке:")
print(f"Вариант 7: accuracy = {test_acc_r7:.4f}")
print(f"Базовый классификатор: accuracy = {test_acc_rbase:.4f}")

# Графики
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history_r7.history['loss'], label='V7 train loss')
plt.plot(history_r7.history['val_loss'], label='V7 val loss')
plt.plot(history_rbase.history['loss'], '--', label='Base train loss')
plt.plot(history_rbase.history['val_loss'], '--', label='Base val loss')
plt.title('Функция потерь (Reuters)')
plt.legend()
plt.subplot(1,2,2)
plt.plot(history_r7.history['accuracy'], label='V7 train acc')
plt.plot(history_r7.history['val_accuracy'], label='V7 val acc')
plt.plot(history_rbase.history['accuracy'], '--', label='Base train acc')
plt.plot(history_rbase.history['val_accuracy'], '--', label='Base val acc')
plt.title('Точность (Reuters)')
plt.legend()
plt.tight_layout()
plt.show()

# Пример предсказания
sample_idx = 5
sample = test_features_r[sample_idx:sample_idx+1]
pred_cat = reuters_v7.predict(sample)
pred_class = int(np.argmax(pred_cat))
true_class = test_labels_r[sample_idx]
print(f"\nПример тестовой новости {sample_idx}:")
print(f"Истинная тема (индекс): {true_class}")
print(f"Предсказанная тема (индекс): {pred_class}")

# Этап 3. Регрессия (Boston Housing)
# Датасет Boston Housing удалён из tf.keras, но доступен через tensorflow_datasets или альтернативно.
print("ЭТАП 3. Регрессия цен на жильё (Boston Housing)")

try:
    (x_train_b, y_train_b), (x_test_b, y_test_b) = keras.datasets.boston_housing.load_data()
except Exception:
    # Альтернативная загрузка из файлов
    import wget, os, gzip, shutil
    url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
    filename = "housing.data"
    if not os.path.exists(filename):
        wget.download(url, filename)
    data = np.loadtxt(filename)
    x_all = data[:, :-1]
    y_all = data[:, -1]
    # Разделим случайно, но с фиксированным seed для воспроизводимости
    np.random.seed(42)
    indices = np.random.permutation(len(x_all))
    split = int(0.8 * len(x_all))
    train_idx, test_idx = indices[:split], indices[split:]
    x_train_b, y_train_b = x_all[train_idx], y_all[train_idx]
    x_test_b, y_test_b = x_all[test_idx], y_all[test_idx]

# Нормализация признаков
mean = x_train_b.mean(axis=0)
std = x_train_b.std(axis=0)
x_train_b_norm = (x_train_b - mean) / std
x_test_b_norm = (x_test_b - mean) / std

# Модель по варианту 7: 2 слоя 64-32, активации elu-relu
def create_boston_v7():
    inputs = keras.Input(shape=(13,))
    x = layers.Dense(64, activation='elu')(inputs)
    x = layers.Dense(32, activation='relu')(x)
    outputs = layers.Dense(1)(x)
    model = keras.Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
    return model

# Базовая модель: 2 слоя 64-64, активации relu-relu
def create_boston_baseline():
    inputs = keras.Input(shape=(13,))
    x = layers.Dense(64, activation='relu')(inputs)
    x = layers.Dense(64, activation='relu')(x)
    outputs = layers.Dense(1)(x)
    model = keras.Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
    return model

# Кросс-валидация по K блокам (K = кол-во слоёв + 1 = 3)
from sklearn.model_selection import KFold
k = 3
kfold = KFold(n_splits=k, shuffle=True, random_state=42)

def cross_val_score(model_creator, X, y, kf):
    mae_scores = []
    for train_idx, val_idx in kf.split(X):
        X_train_k, X_val_k = X[train_idx], X[val_idx]
        y_train_k, y_val_k = y[train_idx], y[val_idx]
        model = model_creator()
        model.fit(X_train_k, y_train_k, epochs=100, batch_size=32, verbose=0)
        _, mae = model.evaluate(X_val_k, y_val_k, verbose=0)
        mae_scores.append(mae)
    return np.mean(mae_scores)

print("Кросс-валидация (3 блока) ")
mean_mae_v7 = cross_val_score(create_boston_v7, x_train_b_norm, y_train_b, kfold)
mean_mae_base = cross_val_score(create_boston_baseline, x_train_b_norm, y_train_b, kfold)
print(f"Вариант 7: средний MAE = {mean_mae_v7:.2f} (тыс.)")
print(f"Базовый классификатор: средний MAE = {mean_mae_base:.2f} (тыс.)")

# Обучение финальной модели на всех обучающих данных и проверка на тесте
print("Обучение финальной модели на всех обучающих данных")
final_v7 = create_boston_v7()
history_boston = final_v7.fit(x_train_b_norm, y_train_b, epochs=200, batch_size=32, verbose=0)
test_loss, test_mae = final_v7.evaluate(x_test_b_norm, y_test_b, verbose=0)
print(f"Вариант 7: тестовый MAE = {test_mae:.2f} (тыс.)")

# График обучения
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history_boston.history['loss'], label='MSE loss')
plt.title('Функция потерь (MSE) – Boston')
plt.legend()
plt.subplot(1,2,2)
plt.plot(history_boston.history['mae'], label='MAE (тыс.)')
plt.title('Средняя абсолютная ошибка (MAE)')
plt.legend()
plt.tight_layout()
plt.show()

# Пример предсказания
sample_idx = 10
x_sample = x_test_b_norm[sample_idx:sample_idx+1]
true_price = y_test_b[sample_idx]
pred_price = final_v7.predict(x_sample, verbose=0)[0,0]
print(f"\nПример тестового дома {sample_idx}:")
print(f"Истинная цена: {true_price:.2f} тыс.")
print(f"Предсказанная цена: {pred_price:.2f} тыс.")
print("Лабораторная работа №4 (вариант 7) успешно выполнена")
```
